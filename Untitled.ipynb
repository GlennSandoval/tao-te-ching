```python
import string
import re

with open('Ursula K Le Guin.md', 'r') as f:
    tao_te_ching_md_raw = f.read()
```

```python
tao_te_ching_md_raw
```

```python
tao_te_ching_md = map(string.strip, tao_te_ching_md_raw.split('\n'))

list(zip(range(10),tao_te_ching_md[:10]))
```

```python
chapter_heading_pattern = re.compile(r'# (?P<chapter>[\d]{1,2})[ ]*')
print re.search(chapter_heading_pattern, tao_te_ching_md[5]).groupdict()['chapter']
chapter_name_pattern = re.compile(r'##[ ]*(?P<chapter_name>(?:[ ]*\w+)*)[ ]*')
print re.search(chapter_name_pattern, tao_te_ching_md[6]).groupdict()['chapter_name']
filter(lambda line: re.search(chapter_heading_pattern, line), tao_te_ching_md)
```

```python
chapter_pattern = re.compile(r'# (?P<chapter>[\d]{1,2})[ ]*\n##[ ]*(?P<chapter_name>(?:[ ]*\w+)*)[ ]*\n')
chapter_matches = re.findall(chapter_pattern, tao_te_ching_md_raw)
[match for match in chapter_matches]
```
